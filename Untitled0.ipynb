```python
import tensorflow as tf
import tensorflow.keras.layers as tfl
import keras.backend as k
```

```python
def create_model():
    #inputs = tfInput((64, 64, 1))
    inputs=tf.keras.Input((64,64,1))
    x = tfl.Conv2D(96, (11, 11), padding="same", activation="relu")(inputs)
    x = tfl.MaxPooling2D(pool_size=(2, 2))(x)
    x = tfl.Dropout(0.3)(x)

    x = tfl.Conv2D(256, (5, 5), padding="same", activation="relu")(x)
    x = tfl.MaxPooling2D(pool_size=(2, 2))(x)
    x = tfl.Dropout(0.3)(x)

    x = tfl.Conv2D(384, (3, 3), padding="same", activation="relu")(x)
    x = tfl.MaxPooling2D(pool_size=(2, 2))(x)
    x = tfl.Dropout(0.3)(x)

    pooledOutput = tfl.GlobalAveragePooling2D()(x)
    pooledOutput = tfl.Dense(1024)(pooledOutput)
    outputs = tfl.Dense(128)(pooledOutput)

    model = tf.keras.Model(inputs,outputs)
    return model
```

```python
feature_extractor = create_model()
imgA = tf.keras.Input(shape=(64, 64, 1))
imgB = tf.keras.Input(shape=(64,64,1))
featA = feature_extractor(imgA)
featB = feature_extractor(imgB)
def euclidean_distance(vectors):
    (featA, featB) = vectors
    sum_squared = k.sum(k.square(featA - featB), axis=1, keepdims=True)
    return k.sqrt(k.maximum(sum_squared, k.epsilon()))

distance = tfl.Lambda(euclidean_distance)([featA, featB])
outputs = tfl.Dense(1, activation="sigmoid")(distance)
model = tf.keras.Model(inputs=[imgA, imgB], outputs=outputs)
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
data = fetch_olivetti_faces()
images=data.images
#plt.matshow(images[0])
labels=data.target

```

```python
import numpy as np
```

```python
def generate_train_image_pairs(images_dataset, labels_dataset):
    unique_labels = np.unique(labels_dataset)
    label_wise_indices = dict()
    for label in unique_labels:
        label_wise_indices.setdefault(label,
                                      [index for index, curr_label in enumerate(labels_dataset) if
                                       label == curr_label])

    pair_images = []
    pair_labels = []
    for index, image in enumerate(images_dataset):
      pos_indices = label_wise_indices.get(labels_dataset[index], [])
      if len(pos_indices) == 0:
          continue
      pos_image = images_dataset[np.random.choice(pos_indices)]
      pair_images.append((image, pos_image))
      pair_labels.append(1)

      neg_indices = [idx for idx in range(len(labels_dataset)) if idx != index and labels_dataset[idx] != labels_dataset[index]]
      if len(neg_indices) == 0:
        continue
      neg_image_index = np.random.choice(neg_indices)
      neg_image = images_dataset[neg_image_index]
      pair_images.append((image, neg_image))
      pair_labels.append(0)

    return np.array(pair_images), np.array(pair_labels)
```

```python
pair_images, pair_labels = generate_train_image_pairs(images, labels)
```

```python
num_pairs_to_display = 5

for i in range(num_pairs_to_display):
    image_pair = pair_images[i]
    image1, image2 = image_pair
    label_pair=pair_labels[i]

    print(label_pair)

    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(image1, cmap='gray')
    plt.title('Image 1')


    plt.subplot(1, 2, 2)
    plt.imshow(image2, cmap='gray')
    plt.title('Image 2')

    plt.tight_layout()
    plt.show()




```

```python
history = model.fit([pair_images[:, 0], pair_images[:, 1]], pair_labels[:],validation_split=0.1,batch_size=64,epochs=100)
```

```python
training_accuracy = history.history['accuracy']
validation_accuracy = history.history['val_accuracy']
epochs = range(1, len(training_accuracy) + 1)

plt.plot(epochs, training_accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, validation_accuracy, 'r', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
def generate_test_image_pairs(images_dataset, labels_dataset, image):
    unique_labels = np.unique(labels_dataset)
    label_wise_indices = dict()
    for label in unique_labels:
        label_wise_indices.setdefault(label,
                                      [index for index, curr_label in enumerate(labels_dataset) if
                                      label == curr_label])

    pair_images = []
    pair_labels = []
    for label, indices_for_label in label_wise_indices.items():
        test_image = images_dataset[np.random.choice(indices_for_label)]
        pair_images.append((image, test_image))
        pair_labels.append(label)
    return np.array(pair_images), np.array(pair_labels)
```

```python
image = images[92] # a random image as test image
test_image_pairs, test_label_pairs = generate_test_image_pairs(images, labels, image) # produce an array of test image pairs and test label pairs

# for each pair in the test image pair, predict the similarity between the images
for index, pair in enumerate(test_image_pairs):
    pair_image1 = np.expand_dims(pair[0], axis=-1)
    pair_image1 = np.expand_dims(pair_image1, axis=0)
    pair_image2 = np.expand_dims(pair[1], axis=-1)
    pair_image2 = np.expand_dims(pair_image2, axis=0)
    prediction = model.predict([pair_image1, pair_image2])[0][0]
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(pair[0], cmap='gray')
    plt.title('Image 1')

    plt.subplot(1, 2, 2)
    plt.imshow(pair[1], cmap='gray')
    plt.title('Image 2')

    plt.tight_layout()
    plt.show()

    print(f"Predicted Similarity: {prediction}")
```
